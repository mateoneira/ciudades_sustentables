### Listas
Una lista es una secuencia ordenada y mutable de elementos. Los elementos de una lista son objetos arbitrarios. Por 'mutable' queremos decir que podemos cambiar los valores dentro de la lista, y por 'ordenada' queremos decir que cada elemento está indexado y se puede acceder a él a través de este índice.
> milista = []

````python
# las listas se definen con [] o list()
lista1=[1,2,3,4,5,6]
print(lista1)
# las listas se pueden rebanar, observa aquí que el número entre [] se refiere al índice
print (lista1[1])
print (lista1[2:])

# métodos de lista - pop y append
lista1=[1,2,3,4,5,6]

#podemos añadir un elemento al final de una lista con .append
lista1.append(7)

#podemos eliminar un elemento de la lista con .pop
lista1.pop(2)
print (lista1)
````

#### Sentencia if
La sentencia condicional en Python se maneja con la sentencia IF.

````python
### condiciones if/else
pasos=9
if pasos==10:
    print ("Verdadero")
else:
    print ("Falso")
````

#### Sentencia for
Las sentencias for se usan cuando tienes un bloque de código que quieres repetir. La palabra clave **For** se usa para iterar sobre un objeto iterable como cadenas, listas, rangos.

````python
# lista como un iterador
lista1=[1,2,3,4,5,6]
for i in lista1:
    print (i)
````

#### Sentencia while

La sentencia while se usa cuando una condición necesita ser verificada en cada iteración, o para repetir un bloque de código para siempre.

````python
### bucles while
pasos=10
i=0
while i<pasos:
    print (f'hola mundo: {i}')
    i=i+1
````

### Clases: objetos en python

Todo es un objeto en Python que se define usando clases. Estas se definen con la palabra clave class.

> class miClase:

Las funciones dentro de los objetos se llaman métodos. Podemos acceder a estos con un punto.

> "hola mundo!".upper()

````python
class Ciudad:
    def __init__(self, nombre, pob, area):
        self.nombre = nombre
        self.poblacion = pob
        self.area = area
        self.densidad = pob/area
    def info(self):
        print(f'nombre: {self.nombre}, poblacion: {self.poblacion}, area: {self.area}, densidad {self.densidad}')
````

````python
londres = Ciudad('Londres', 8961989, 1572)
````

````python
londres.info()
````

### Subconjunto de datos

Una funcionalidad importante es poder cortar, trocear y crear subconjuntos de nuestros datos. Esto se puede hacer de múltiples maneras.

La indexación más básica en pandas es []:
> datos['columna']

Se puede realizar una indexación más avanzada usando _.loc_

* _.loc_ permite el uso de etiquetas para obtener nuestros datos
*_.iloc_ permite el uso de la posición entera para subconjuntar los datos

````python
#usa la indexación básica [] para obtener los datos de la columna sub_region_1
data['sub_region_1']
````

También podemos usar operadores booleanos para seleccionar solo los datos que cumplen un criterio específico

````python
#aquí seleccionamos solo los datos donde sub_region_1 es igual a 'Greater London'
london_data = data[data['sub_region_1']=='Greater London' ]
````

````python
# usa .loc para obtener datos por la etiqueta del índice
london_data.loc[37164]
````

````python
# también podemos usar .loc para obtener una columna específica de la siguiente manera
london_data.loc[:,'retail_and_recreation_percent_change_from_baseline']
````

````python
#o podemos usar una lista de las columnas que queremos subconjuntar
columns = ['retail_and_recreation_percent_change_from_baseline', 
           'grocery_and_pharmacy_percent_change_from_baseline', 
           'transit_stations_percent_change_from_baseline',
           'workplaces_percent_change_from_baseline',
           'residential_percent_change_from_baseline',
           'date'
          ]

#asignar subconjunto de datos a una nueva variable llamada subconjunto
subset = london_data.loc[:,columns]
subset.head()
````

### Graficando con Pandas

Podemos graficar directamente los datos en nuestro dataframe usando el método _.plot()_

````python
#fecha a formato datetime
subset.date = pd.to_datetime(subset.date, infer_datetime_format=True)
#establecer el índice a la fecha
subset.set_index('date', inplace=True, drop=True)
#graficar
subset.plot(figsize= (15,7))
````

## Geopandas

GeoPandas es un proyecto de código abierto para facilitar el trabajo con datos geoespaciales en python. GeoPandas extiende los tipos de datos utilizados por pandas para permitir operaciones espaciales en tipos geométricos. Las operaciones geométricas son realizadas por shapely. Geopandas además depende de fiona para el acceso a archivos y de descartes y matplotlib para la representación gráfica.

> import geopandas as gpd

Podemos leer shapefiles y otros tipos de datos espaciales en python usando geopandas con _.read_file()

````python
import geopandas as gpd
import matplotlib
import matplotlib.pyplot as plt

#ubicación del shapefile
fp = "data/LSOA_2011_London_gen_MHW.shp"

#leer shapefile
gdf=gpd.read_file(fp)
````

````python
#podemos usar todas las funciones disponibles en pandas, como .head()
gdf.head()
````

````python
## es importante tener en cuenta que geopandas siempre tiene una columna de geometría
## aquí las geometrías se almacenan como objetos de shapely, no debemos cambiar
### el nombre de la columna de geometría
print(gdf.columns)
````

````python
# geopandas puede hacer gráficos geográficos con un simple método plot.

# Configurar la figura de matplotlib
matplotlib.rcParams['font.size'] = 12

# establecer a epsg=3857 
gdf = gdf.to_crs(epsg=3857)

# método geodataframe.plot
ax=gdf.plot(figsize=(20, 20), alpha=0.8, linewidth=0.1,edgecolor='w',
            column='POPDEN',legend=True,vmin=1,vmax=200,
           legend_kwds={'label': "Densidad de población (persona/HA)",'orientation': "vertical",'shrink': 0.5})

# esto establece el título
ax.set_title("Densidad de población en el Gran Londres",fontsize= 30)

# esto elimina el eje
ax.set_axis_off()
````
